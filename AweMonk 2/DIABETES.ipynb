```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
```

## Data Processing

```python
#Reading the CSV file
df=pd.read_csv('Data.csv')
df.head()
```

```python
#dopping/Removing the Unnamed column from the DataFrame.
col_names = df.columns
df = df.drop(df.columns[0],axis=1)
for c in col_names:
    df = df.replace("?", np.NaN)
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
df
```

```python
print("Dataset shape:", df.shape)
```

```python
df=df.drop(columns=['desy_sort_key','gender_id','DiseaseType','race','Treated'],axis=1)
```

## Feature Engineering

```python
# A categorical variable creation process is performed according to the value.
def Classlevel(row):
    if ((row["Hemoglobin_A1C"] >= 6.5) and (row["Fasting_Glucose"] >= 126) and (row["Random_Glucose"] >= 200)):
        return "Critical"
    elif ((row["Hemoglobin_A1C"] >= 5.7 and row["Hemoglobin_A1C"] <= 6.4) and (row["Fasting_Glucose"] >= 100 and row["Fasting_Glucose"] <= 125) ):
        return "Early on set"
    else:
        return "No Risk"
```

```python
# The operation performed was added to the dataframe.
df = df.assign(ClassLevel=df.apply(Classlevel, axis=1))

df
```

```python
df['ClassLevel'].value_counts()
```

## Label Encoding

```python
# Importing LabelEncoder from Sklearn
# library from preprocessing Module.
from sklearn.preprocessing import LabelEncoder
 
# Creating a instance of label Encoder.
le = LabelEncoder()
 
# Using .fit_transform function to fit label
# encoder and return encoded label
label = le.fit_transform(df['ClassLevel'])
 
# printing label
label
```

```python
# removing the column 'Purchased' from df
# as it is of no use now.
df.drop("ClassLevel", axis=1, inplace=True)
 
# Appending the array to our dataFrame
# with column name 'Purchased'
df["ClassLevel"] = label
 
# printing Dataframe
df
```

## Splitting the Features and ClassLevel

```python
X=df.drop(columns='ClassLevel',axis=1)
y=df['ClassLevel']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,stratify=y, random_state = 2)
```
